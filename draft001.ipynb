## How many businesses are in Koreatown?

This dataset contains all currently active businesses in Los Angeles, updated mid-January 2019. 

More info [here](https://data.lacity.org/A-Prosperous-City/Listing-of-Active-Businesses/6rrh-rzua); from https://data.lacity.org

Columns in this dataset:
```
'LOCATION ACCOUNT #'
'BUSINESS NAME'
'DBA NAME'
'STREET ADDRESS'
'CITY'
'ZIP CODE'
'LOCATION DESCRIPTION'
'MAILING ADDRESS'
'MAILING CITY'
'MAILING ZIP CODE'
'NAICS'
'PRIMARY NAICS DESCRIPTION'
'COUNCIL DISTRICT'
'LOCATION START DATE'
'LOCATION END DATE'
'LOCATION'
```
At first glance: 
* There are 514,878 businesses in this dataset, with 16 columns.
* It includes metropolitan-area listings (i.e.: Glendale, Wilmington(!), Pacific Palisades, Fullerton(!!)...). We just want listings inside Koreatown, within Los Angeles city limits.
* Mailing addresses are included. Columns related to mailing addresses (`'MAILING ADDRESS', 'MAILING CITY', 'MAILING ZIP CODE'`) will not be needed for this analysis.
* The `'LOCATION END DATE'` column appears to be full of `NaN`. As this is a collection of active businesses, this column, being unnecessary, will also be dropped.
* NAICS appears to be a Dewey Decimal System-like categorization scheme for businesses.
* Column names will need to be converted to Pythonic style for ease of use.
* This collection also includes single family housing constructions (NAICS code 233210).
* Geolocations accompany some of the listings (in the `'LOCATION'` column).

Additional facts:
* Many of Los Angeles's neighborhoods (mostly in the San Fernando Valley) are treated like separate cities by the USPS and residents. Woodland Hills, Hollywood, Northridge, Studio City, _et al_. all lie within Los Angeles city limits. Not relevant to this particular analysis, but useful for understanding composition of Los Angeles neighborhoods.

Out of 514,878 listed businesses, just under half are located within city limits and outside the San Fernando Valley. We can narrow this down further by ZIP code to obtain just Koreatown listings:

````python
# zip+4 information is repeated in 'location_description'
# column, so not needed here

la_biz['zip_code'] = (la_biz['zip_code']
                      .str.split(n=1, pat='-', expand=True)
                      .iloc[:,0])
````

````python
# all listings in koreatown, by zip code (90010, 90005, 90006, acc. to wikipedia)
# will need refining later

ktown_biz = la_biz.loc[((la_biz['city'] == 'LOS ANGELES')
                        & ((la_biz['zip_code'] == '90005')
                           | (la_biz['zip_code'] == '90006')
                           | (la_biz['zip_code'] == '90004')
                           | (la_biz['zip_code'] == '90010')
                           | (la_biz['zip_code'] == '90020') 
                           | (la_biz['zip_code'] == '90019'))
                       ), :]
````

````python
ktown_biz.shape
````

Within these ZIP codes, Koreatown contains 32309 businesses.

## How many Korean BBQ restaurants are in Koreatown?

Upon closer investigation, not all Koreatown Korean barbecue restaurants are called BBQ: 
* 'Galbi/Galbee/Kalbi' and 'Bulgogi' are commonly found in the names of Korean BBQ restaurants. _Galbi_ means 'ribs' and _bulgogi_ refers to marinated meats.
* Many such restaurants are named for their specialty, i.e.: Ahgassi Gopchang (_gopchang_ meaning 'small intestines').
* One establishment is called 'Ssam'. _Ssam_ is the word for lettuce or soaked rice paper wrapped around grilled meats.
* At least one restaurant devoted to Korean BBQ is called 'Barbeque'.
* Soot Bull Jeep and The Corner Place are both Korean BBQ restaurants in K-town that have BBQ nowhere in the name.

The NAICS codes will not be as helpful as I thought for restaurants, as many are miscategorized or lack a NAICS code entirely.

````python
ktown_bbq = ktown_biz.loc[(ktown_biz['dba_name'].str.contains('BBQ|BARBEQUE|GALBI|GALBEE|SSAM|BULGOGI|GOPCHANG|GOBCHANG') == True) 
                          | (ktown_biz['business_name'].str.contains('BBQ|BARBEQUE|GALBI|GALBEE|SSAM|BULGOGI|GOPCHANG|GOBCHANG') == True), :]
ktown_bbq
````

````python
ktown_bbq.shape
````

This query yields a few false positives, as well as great Korean restaurants that don't serve barbeque, so we will sift those out:

````python
ktown_bbq = ktown_bbq.drop([186826,189188,201927,237189,321223,332246,341183,392750,409755,431022,482262,429652,481775], axis=0)
````
